```python
from sklearn import datasets
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X_iris = iris.data[:, [2, 3]]
y_iris = iris.target
```

1. sepal length in cm
2. sepal width in cm
3. petal length in cm

- Iris Versicolour
- Iris Virginica

```python
y_iris
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=y_iris)
plt.show()
```

```python
data_X = X_iris[y_iris != 0]
data_X
```

```python
data_Y = y_iris[y_iris != 0]
```

```python
data_Y
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(data_X, data_Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
from matplotlib.colors import ListedColormap

# DecisionBoundaryDisplay.from_estimator(model, data_X)

cmap_light = ListedColormap(["orange", "cyan", "cornflowerblue"])

fig, ax = plt.subplots(figsize=(10, 7))
ax.scatter(data_X[:, 0], data_X[:, 1], c=data_Y)
DecisionBoundaryDisplay.from_estimator(
    model,
    data_X,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    cmap=cmap_light,
    alpha=0.5)
```

```python
import sklearn
sklearn.__version__
```

```python
from sklearn.metrics import confusion_matrix

pred_Y = model.predict(data_X)

confusion_matrix(data_Y, pred_Y)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay.from_estimator(model, data_X, data_Y)
```
